# Data Cleaning for Past DSA3101 Data (Data for 4 Semesters)

```python
# load necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data for 2120
data_2120 = pd.DataFrame(pd.read_excel("for_anonymising/2120/DSA3101 Project Preferences_February 27, 2022_17_edit.43_shared.xlsx"))
result_2120 = pd.DataFrame(pd.read_excel("for_anonymising/2120/student-preferences_shared.xlsx"))

# load data for 2210
data_2210 = pd.DataFrame(pd.read_excel("for_anonymising/2210/DSA3101-2210+Project+Preferences_February+13,+2025_11.01_shared.xlsx"))
result_2210 = pd.DataFrame(pd.read_excel("for_anonymising/2210/DSA3101-2210_project_allocations_canvas_shared.xlsx"))

# load data for 2220
data_2220 = pd.DataFrame(pd.read_excel("for_anonymising/2220/DSA3101-2220+Project+Preferences_February+13,+2025_11.02_shared.xlsx"))
result_2220 = pd.DataFrame(pd.read_excel("for_anonymising/2220/proj_pref_upload_shared.xlsx"))

# load data for 2310
data_2310 = pd.DataFrame(pd.read_excel("for_anonymising/2310/DSA3101-2310+Project+Preferences_February+13,+2025_11.03_shared.xlsx"))
result_2310 = pd.DataFrame(pd.read_excel("for_anonymising/2310/groupings_with_ic_shared.xlsx"))
```

```python
result_2120['student_id'] = range(len(result_2120))
result_2120
```

```python
# anonymization of student ID
mapping_dict = dict(zip(result_2120["User ID"], result_2120["student_id"]))
group_columns = ["Group members_1", "Group members_2", "Group members_3", "Group members_4", "Individual_1"]
data_2120[group_columns] = data_2120[group_columns].applymap(lambda x: mapping_dict.get(x, x) if pd.notna(x) else x)

# formulating is_selfformed_grp binary indicator from Q1 column
self_formed_grp_mapping = {
        'I am on my own, and looking for a group to join.': 0,
        'I have a group (at least 2 including myself) of people I would like to work with.': 1
    }

data_2120['is_self_formed_grp'] = data_2120['Q1'].replace(self_formed_grp_mapping) 
data_2120 = data_2120.drop(columns = "Q1", inplace = False)
data_2120 = data_2120.dropna(subset=group_columns, how = "all") #drop rows where all values in group columns are NaN
data_2120
```
