# Data Cleaning for Past DSA3101 Data (Data for 4 Semesters)

```python
# load necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data for 2120
data_2120 = pd.DataFrame(pd.read_excel("for_anonymising/2120/DSA3101 Project Preferences_February 27, 2022_17_edit.43_shared.xlsx"))
result_2120 = pd.DataFrame(pd.read_excel("for_anonymising/2120/student-preferences_shared.xlsx"))

# load data for 2210
data_2210 = pd.DataFrame(pd.read_excel("for_anonymising/2210/DSA3101-2210+Project+Preferences_February+13,+2025_11.01_shared.xlsx"))
result_2210 = pd.DataFrame(pd.read_excel("for_anonymising/2210/DSA3101-2210_project_allocations_canvas_shared.xlsx"))

# load data for 2220
data_2220 = pd.DataFrame(pd.read_excel("for_anonymising/2220/DSA3101-2220+Project+Preferences_February+13,+2025_11.02_shared.xlsx"))
result_2220 = pd.DataFrame(pd.read_excel("for_anonymising/2220/proj_pref_upload_shared.xlsx"))

# load data for 2310
data_2310 = pd.DataFrame(pd.read_excel("for_anonymising/2310/DSA3101-2310+Project+Preferences_February+13,+2025_11.03_shared.xlsx"))
result_2310 = pd.DataFrame(pd.read_excel("for_anonymising/2310/groupings_with_ic_shared.xlsx"))
```

```python
# 1) Gather all IDs from data_2120 columns of interest
columns_of_interest = ["Group members_1", "Group members_2",
                       "Group members_3", "Group members_4", "Individual_1"]
unique_ids_data = pd.unique(data_2120[columns_of_interest].stack().dropna())

# 2) Gather all IDs from result_2120 "User ID" column
unique_ids_result = pd.unique(result_2120["User ID"].dropna())

# 3) Combine them so we build ONE mapping for both tables
all_ids = np.unique(np.concatenate([unique_ids_data, unique_ids_result]))

# 4) Build a single anonymization mapping
mapping_dict = {
    old_id: f"student_{i+1}" for i, old_id in enumerate(all_ids)
}

# 5) Define a helper function to apply the mapping
def anonymize_id(x):
    if pd.isna(x):
        return x
    else:
        return mapping_dict.get(x, x)  # default to x if not found in mapping

# 6) Apply the mapping to data_2120 columns of interest
data_2120[columns_of_interest] = data_2120[columns_of_interest].applymap(anonymize_id)

# 7) Apply the mapping to result_2120 "User ID"
result_2120["User ID"] = result_2120["User ID"].apply(anonymize_id)
```

```python
# Columns in data_2120 that contain student IDs
columns_of_interest = ["Group members_3", "Group members_4", "Individual_1"]

# 1. Collect all unique IDs from data_2120
all_ids = set()
for col in columns_of_interest:
    # dropna() to skip NaN
    col_ids = data_2120[col].dropna().unique()
    all_ids.update(col_ids)

# 2. Collect all unique IDs from result_2120
all_ids.update(result_2120["User ID"].dropna().unique())

# Sort for stable, repeatable order
sorted_ids = sorted(all_ids)

# Build the anonymization dictionary
anonymize_dict = {}
for i, original_id in enumerate(sorted_ids, start=1):
    anonymize_dict[original_id] = f"student_{i}"

# 3A. Apply to data_2120
for col in columns_of_interest:
    data_2120[col] = data_2120[col].map(anonymize_dict)

# 3B. Apply to result_2120
result_2120["User ID"] = result_2120["User ID"].map(anonymize_dict)

result_2120["User ID"] = result_2120["User ID"].map(anonymize_dict).fillna(result_2120["User ID"])
result_2120
```
